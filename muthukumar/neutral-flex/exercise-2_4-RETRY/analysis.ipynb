```python
import os

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
filenames = [f'msd-files/{file}' for file in os.listdir(os.path.join(os.getcwd(), 'msd-files')) if '_V2' not in file]
data = np.array([np.loadtxt(fname, skiprows=1) for fname in filenames])
```

```python
idx = [20, 60, 100, 'CM']
```

```python
fig, ax = plt.subplots(dpi=300)

for i, a in enumerate(data):
    ax.plot(a[:, 0] * 5000 * 1e-3, a[:, 1], linewidth=0.75, label=f'i = {idx[i]}')
    
ax.plot([2e2, 1e3], np.array([2e2, 1e3]) ** 0.6 + 10, 'k')
ax.text(2e2, 60, 'slope = 0.6', fontsize='x-small')
ax.plot([3e4, 1e5], np.array([3e4, 1e5]) ** 1, 'k')
ax.text(3e4, 5e4, 'slope = 1', fontsize='x-small')

ax.set_xlim(left=1e2, right=2e5)
ax.set_ylim(bottom=1)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('$\\tau$')
ax.set_ylabel(r'$ \langle [R_{i}(t + \tau) - R_{i}(t)]^2 \rangle $')
ax.set_title('N = 120')
ax.legend()

fig.tight_layout()
```

```python
filenames = [f'msd-files/{file}' for file in os.listdir(os.path.join(os.getcwd(), 'msd-files')) if '_V2' in file]
data = np.array([np.loadtxt(fname, skiprows=1) for fname in filenames])
```

```python
fig, ax = plt.subplots(dpi=300)

for i, a in enumerate(data):
    ax.plot(a[:, 0] * 5000 * 1e-3, a[:, 1], linewidth=0.75, label=f'i = {idx[i]}')

ax.plot([3e4, 5e4], np.array([3e4, 5e4]) ** 0.5 + 3e3, 'k')
ax.text(3e4, 4e3, 'slope = 0.5', fontsize='x-small')
ax.plot([7e4, 1e5], np.array([7e4, 1e5]) ** 1, 'k')
ax.text(7e4, 5e4, 'slope = 1', fontsize='x-small')

ax.set_xlim(left=1e3, right=2e5)
ax.set_ylim(bottom=10)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('$\\tau$')
ax.set_ylabel(r'$ \langle [R_{i}(\tau) - R_{i}(0)]^2 \rangle $')
ax.set_title('N = 120')
ax.legend()

fig.tight_layout()
```
